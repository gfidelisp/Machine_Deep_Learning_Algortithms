## Importing the libraries

```python
import pandas as pd
import copy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sn

from matplotlib import rc

import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning)

rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

## Functions

#### Activation function

```python
def activation(z, act):
    """
    Activation function
    
    activation (z, act)    
    
    return (func) 
    """

    # Sigmoid
    if act == 'sig':
        func = 1/(1+np.exp(-z))

    # Hiperbolic tangent
    elif act == 'tanh':
        func = np.tanh(z)
    
    # Rectified Linear Unit (ReLU)
    elif act == 'Relu':
        func = np.maximum(0,z)
    
    else:
        print('Error! Activation function is not identied.')

    return func
```

#### Derivative of the ReLU function

```python
def derivates(z, act, X):
    """
    Função das derivadas
    
    derivates(a, act):

    return(der)
    """
   
    # Função sigmoide
    if act == 'sig':
        a = activation(z, act)
        a = np.vstack((a,np.ones(len(X))))
        der = a*(1-a)


    # Função tangente hiperbólica
    elif act == 'tanh':
        a = activation(z, act)
        a = np.vstack((a,np.ones(len(X))))
        der = (1-a**2)

    
    # Função ReLU
    elif act == 'Relu':
        z[z<=0] = 0
        z[z>0] = 1

        der = np.vstack((z,np.ones(len(X))))

    else:
        print('Erro! Função de Ativação não identificada.')

    return der  
```

#### Função Custo - Binary Cross-Entropy

```python
def BCE(X,W,y,Neurons, act):
    """
    
    Função  Custo 
    
    BCE(X,W,y,Neurons)
    
    return -np.sum(y*np.log(h) + (1-y)*np.log(1-h))/len(y)
    
    """
    
    # Cálculo da saída da Rede Neural
    a_mat = forward_prop(X, W, Neurons, act)
    h = a_mat[-1]
    
    return -np.sum(y*np.log(h) + (1-y)*np.log(1-h))/len(y)
```

#### Função para o Cálculo Gradiente Numericamente

```python
def Grad(W, X, y, Neurons):
    
    """
    
    Gradiente aproximado
    
    rad(W, X, y, Neurons)
    
    return Grad_aprox
    
    """
    
    # Incremento nos pesos e vieses
    Epsilon = 10**-15
    
    # Copiando o tamanho do vetor de pesos e vieses
    Grad_aprox = copy.deepcopy(W)
    
    #Iterando em cada uma das camadas, nós de entrada e nós de saída
    for i in range(len(W)):
        for j in range(np.shape(W[i])[0]):
            for k in range(np.shape(W[i])[1]):
                
                # Copiando duas listas para oos valores positivos e negativos
                W_plus = copy.deepcopy(W)
                W_minus = copy.deepcopy(W)
                
                # Adicionando os incrementos em elementos selecionados das listas
                W_plus[i][j][k] = W_plus[i][j][k] + Epsilon
                W_minus[i][j][k] = W_minus[i][j][k] - Epsilon
                
                #Cálculo do Gradiente
                Grad_aprox[i][j][k] = (BCE(X, W_plus, y, Neurons) - BCE(X, W_minus, y, Neurons))/(2*Epsilon)
            
    return Grad_aprox
```

#### Forward Propagation

```python
def forward_prop(X, W, Neurons, act):
    """
    
    Função Forward Propagation
    
    forward_prop(X, W)
    
    return a_mat
    
    """
    
    # a^(1) = X + 1
    a = np.vstack((np.array(X).T, np.ones(len(np.array(X)))))
    
    # Guardando o valor de a no vetor a_vec
    a_vec = copy.deepcopy(a)
    
    #Iterando em cada camada
    for c,v in enumerate(W):

        # Coletando os dados do vetor a 
        a = a_vec[-np.shape(v)[1]:]

        # z^(c) = theta^(c-1)*a^(c-1)
        z = np.dot(v,a)

        # Para toda a camada, exceto a última, adicionar um vetor 1 para a:
        if c != len(W)-1:            
            
            # Aplicando a Função de ativação 
            a = activation(z, act)
            
            # Vetor de elementos 1 a ser adicionado em a
            add_one = np.ones(np.shape(a_vec[-np.shape(v)[1]:])[1])

            # a^(c) = a^(c) + 1        
            a = np.vstack((a,add_one))
        
        else:
            # a^(c) = sigmoid(z^(c))
            a = activation(z, 'sig')
        
        # Guardando o valor de a no vetor a_vec
        a_vec = np.vstack((a_vec,a))
    
    # Transformação da estrtura de a_vec
    a_mat = []
    for i in range(len(Neurons)):
        a_mat.append(a_vec[sum(Neurons[:i])+i:sum(Neurons[:i+1])+i+1])
    
    return a_mat
```

#### Backward Propagation

```python
def backward_prop(a_mat, sigma, Delta, Neurons, y, W, act, X):
    """
    
    Função Backward Propagation
    
    forward_prop(a_mat, sigma, Delta, Neurons, y, W)v
    
    return Delta
    
    """   
    
    # sigma^(L) = a^(L) - y
    sigma[len(Neurons)-2] = (a_mat[-1] - y)
    
    # sigma^(L-1) = W^(L-1)^T*sigma^(L)*(a^(L-1))*(1- a^(L-1))
    i = len(Neurons)-3
    sigma[i] = np.dot(W[i+1].T,sigma[i+1])*derivates(np.dot(W[i],a_mat[i]), act, X)

    # Iterando até sigma^(2)
    for i in range(len(W)-3,-1,-1):

        # sigma^(l-1) = W^(l-1)^T*sigma^(l)*(a^(l-1))*(1- a^(l-1))
        sigma[i] = np.dot(W[i+1].T,sigma[i+1][:-1])*derivates(np.dot(W[i],a_mat[i]), act, X)
    
    # Iterando até L-1
    for i in range(len(sigma)-1):
        
        # Delta^(l) = Delta^(l) + a^(l)*sigma^(l+1) 
        Delta[i] = np.dot(a_mat[i],sigma[i].T).T[:-1]/len(y)
    
    return Delta
```

#### Função de Aprendizado 

```python
def Neural_Network(alpha, it, W, X, sigma, Delta, y, Neurons, X_t, y_t, act, Silent = True, Grad_aprox = False):
    """
    
    Função de aprendizado
        
    Neural_Network(alpha, it, W, X, sigma, Delta, y, Neurons):
    
    print(BCE(X_train,W,y_train.values))
    
    return(W)
    """
    
    Perf = []
    # Iterando até um número de epísódios it    
    for iter in range(it):
        
        # Forward Propagation
        a_mat  = forward_prop(X, W, Neurons, act)
        
        # Backward Propagation
        Del = backward_prop(a_mat, sigma, Delta, Neurons, y, W, act, X)
        
        # Atualizando o valor dos pesos
        W = list(W - alpha*np.array(Del))
        
        # Impriminto a função custo ou acurácia
        acc      = np.sum(np.round(forward_prop(X, W, Neurons, act)[-1]) == y)/len(y)*100
        acc_test = np.sum(np.round(forward_prop(X_t, W, Neurons, act)[-1]) == y_t)/len(y_t)*100
        
        Loss      = BCE(X  , W, y  , Neurons, act)
        Loss_test = BCE(X_t, W, y_t, Neurons, act)
        
        Perf.append([acc, acc_test, Loss, Loss_test])
        
        if Silent == False:
            print('A acurácia na iteração '     + str(iter) + ' é de ' +str(acc))
            print('A função custo na iteração ' + str(iter) + ' é de ' +str(Loss))
        
        # Comparação entre a solução analítica e numérica do gradiente
        if Grad_aprox == True:
            if iter % (it/5) == 0:
                error = np.abs(np.array(Del) - np.array(Grad(W, X, y, Neurons)))
                error_deriv = [error[i].mean() for i in range(len(Neurons)-1)]
                print('Erro das derivadas: ' +str(error_deriv))
    
    accu = np.round(np.sum(np.round(forward_prop(X, W, Neurons, act)[-1]) == y)/len(y)*100,2)
    print('A acurácia do modelo no conjunto de treino é de ' + str(accu) + '%')
    
    cm = confusion_matrix(list(np.round(forward_prop(X_t, W, Neurons, Actv)[-1])[0]),list(y_t))
    
    return(W , Perf, cm)
```

## Funções de Plotagem

```python
def Plot_Acc(Perf, Actv, It, alpha):
    """
    Plot da Acurácia 
    
    
    Plot_Acc(Perf, Actv, It)
    """
    
    plt.figure(figsize = (6, 6/1.618))
    plt.plot(np.array(Perf).T[0],':', label = 'Treino')
    plt.plot(np.array(Perf).T[1],'--', label = 'Teste')

    plt.grid(linestyle = ':')
    plt.legend()
    
    plt.xlabel('Episódio [-]')
    plt.ylabel('Acurácia [\%]')
    
    plt.savefig('Plots/Acurácia_'+str(Actv)+'_'+str(It)+'_' +str(alpha)+'_' + str(Neurons)+'.png', format = 'png',dpi = 600, bbox_inches = 'tight')
```

```python
def Plot_Loss(Perf, Actv, It, alpha):
    """
    Plot do Custo 
    
    
    Plot_Loss(Perf, Actv, It)
    """
    
    plt.figure(figsize = (6, 6/1.618))
    plt.plot(np.array(Perf).T[2],':', label = 'Treino')
    plt.plot(np.array(Perf).T[3],'--', label = 'Teste')
    
    plt.grid(linestyle = ':')
    plt.legend()
    
    plt.xlabel('Episódio [-]')
    plt.ylabel('Função Custo [-]')
    
    plt.savefig('Plots/Custo_'+str(Actv)+'_'+str(It)+'_' +str(alpha)+'_'+ str(Neurons) +'.png', format = 'png',dpi = 600, bbox_inches = 'tight')
```

```python
def Plot_CM(cm):
    
    classes = ["0", "1"]

    df_cfm = pd.DataFrame(cm, index = classes, columns = classes)
    plt.figure(figsize = (6,6/1.618))
    cfm_plot = sn.heatmap(df_cfm, annot=True)
    plt.ylabel('Valor Real')
    plt.xlabel('Valor Predito')
    cfm_plot.figure.savefig('Plots/CM_'+str(Actv)+'_'+str(It)+'_' +str(alpha)+'_'+ str(Neurons)+'.png', format = 'png',dpi = 600, bbox_inches = 'tight')
```

```python
def Plot_Front(W_, Neurons, Actv, It,alpha):
    """
    Plot da Fronteira de Decisão
    
    
    Plot_Loss(Perf, Actv, It)
    """
    x_min, x_max = -1, 1.25
    xx, yy = np.meshgrid(np.linspace(x_min,x_max, 100),np.linspace(x_min, x_max, 100))
    x_in = np.c_[xx.ravel(), yy.ravel()]
    y_pred = np.round(forward_prop(x_in, W_, Neurons, Actv)[-1])
    y_pred = pd.DataFrame(y_pred.reshape(xx.shape))

    df=pd.read_csv("classification2.txt", header=None)

    X=df.iloc[:,:-1].values
    y=df.iloc[:,-1].values


    pos , neg = (y==1).reshape(len(X),1) , (y==0).reshape(len(X),1)
    
    plt.figure(figsize = (6, 6/1.618))
    plt.scatter(X[pos[:,0],0],X[pos[:,0],1],c="b",marker="+")
    plt.scatter(X[neg[:,0],0],X[neg[:,0],1],c="r",marker="o",s=10)
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdBu, alpha=0.5)

    plt.xlabel("x1 [-]")
    plt.ylabel("x2 [-]")
    plt.legend(["1","0"],loc=0)
    plt.savefig('Plots/Fronteiras_'+str(Actv)+'_'+str(It)+'_' +str(alpha)+'_'+ str(Neurons)+'.png', format = 'png',dpi = 600, bbox_inches = 'tight')
```

### Inputs 

```python
Data = pd.read_csv('classification2.txt', names = ['X1', 'X2', 'y'])
```

### Separação dos conjunto de treino, teste e validação

```python
X_train, X_test, y_train, y_test = train_test_split( Data[['X1','X2']], 
                                                     Data['y'], 
                                                     test_size=0.2, 
                                                     random_state=42,
                                                     stratify = Data['y'])
```

## Rede Neural 

```python
# Arquitetura da Rede
Neurons = [X_train.shape[1], 100, 10, 1]

# Taxa de aprendizado
alpha = 1

# Númere de episódios
It = 10000

#Função de Ativação
Actv = 'tanh'

# Estruturas para guardar os dados
# Pesos
W = []

# Derivadas
Delta = []

#Erros
sigma = []


for i in range(len(Neurons)-1):

    # Xavier Initialization
    W.append(np.sqrt(6)/(Neurons[i+1] + Neurons[i])*np.random.randn(Neurons[i+1],Neurons[i]+1))
    
    Delta.append(np.zeros([Neurons[i+1],Neurons[i]+1]))
    sigma.append(np.zeros([1,1]))


# Algoritmo da Rede Neural
W_, Perf, cm = Neural_Network(alpha, It, W, X_train, sigma, Delta, y_train.values, Neurons, 
                          X_test, y_test.values, Actv,Silent = True, Grad_aprox = False)
```

#### Acurácia no conjunto de teste

```python
acc_test = np.round((np.round(forward_prop(X_test, W_, Neurons, Actv)[-1]) == y_test.values).sum()/len(y_test)*100,2)
print('A acurácia do modelo no conjunto de teste é de ' + str(acc_test) + '%')

Plot_Acc(Perf, Actv, It, alpha)
Plot_Loss(Perf, Actv, It, alpha)
Plot_Front(W_, Neurons, Actv, It, alpha)
Plot_CM(cm)
```
