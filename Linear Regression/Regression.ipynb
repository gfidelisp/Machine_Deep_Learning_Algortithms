## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Palatino"
})
```

## Objective function

```python
def f_true(x):
    return 2+0.8*x
```

## Artificial Dataset

```python
xs = np. linspace (-3, 3, 100)
ys = np. array ( [ f_true (x) + np. random . randn () *0.5 for x in xs] )
```

#### Visualizing the Dataset

```python
plt.plot(xs,ys,'o', label = 'Artificial dataset')
plt.plot(xs,f_true(xs),label = 'Original line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.savefig('Plots/Data.png',format = 'png', bbox_inches = 'tight')
```

```python
### Hypothesis function
def h(x,theta):
    return theta [0] + theta[1]*x

### Cost Function
def J(theta, xs, ys):
    return sum((h(xs, theta) - ys)**2)/(2*len(ys))

### Derivative of the cost function
def gradient (i, theta , xs , ys):
    return sum((h(xs, theta) - ys)*xs**i)/(len(ys))

#### Plot function
def print_modelo (theta , xs , ys, grau = 1, i =0):
    plt.figure()
    plt.plot(xs, h(xs,theta), '--', label = 'Prediction')
    plt.plot(xs,ys,'o', label = 'Artificial Data')
    plt.plot(xs,f_true(xs),'-.',label = 'Original Line')
    plt.xlabel('xs')
    plt.ylabel('ys')
    plt.legend()
    plt.grid()
    if i == 0:
        plt.savefig('Linear.png',format = 'png', bbox_inches = 'tight')
    else:
        plt.savefig('Linear_' + str(i)+ '.png',format = 'png', bbox_inches = 'tight')
    pass
```

## Gradient Descent

```python
def grad_desc(alpha, xs, ys,Silent_Val = True, Silent_J = True):

    
    theta = np.zeros(2)
    erro  = 1
    T =np.zeros(2)
    count = 0
    J_count = [J(theta, xs, ys)]
    theta_count = [theta]

    while count < 5000:
        for i in range(len(theta)):
            T[i] = gradient (i, theta , xs , ys)
        theta = theta - alpha*T
        erro = np.linalg.norm(T)
        count = count + 1
        J_count.append(J(theta, xs, ys))
        theta_count.append(theta)


    if Silent_Val == False:
        print_modelo (theta , xs , ys)

        
    if Silent_J == False:
        plt.figure()
        plt.loglog(np.linspace(1,5000,len(J_count)),J_count)
        plt.xlabel('Epoch')
        plt.ylabel('Loss function (J)')
        plt.grid()
        plt.savefig('Plots/J_1.png',format = 'png', bbox_inches = 'tight')
    
    return theta,J_count,theta_count
```

```python
theta,J_count,Theta_count = grad_desc(0.1, xs, ys,Silent_J = False)
```

## Evaluation of the learning rate

```python
alpha = np.array([0.5,0.1,0.0001])
```

```python
J_c = []
for alp in alpha:
    theta,J_count,Theta_count = grad_desc(alp, xs, ys)
    J_c.append(J_count)
```

```python
plt.loglog(np.linspace(1,5000,len(J_count)),J_c[0], 'x--',label = 'Learning rate of 0.5')
plt.loglog(np.linspace(1,5000,len(J_count)),J_c[1], 'o:' , label = 'Learning rate of 0.1')
plt.loglog(np.linspace(1,5000,len(J_count)),J_c[2], 's-.',label = 'Learning rate of 0.0001')
plt.grid()
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Cost function')
plt.savefig('Plots/Cost.png',format = 'png', bbox_inches = 'tight')
```

## Loss function and weights

## Regression of a polynomial curve

## Function

```python
def f_true_pol(x):
    return 2+0.8*x+1.5*x**2
```

## Artificial Dataset

```python
xs = np. linspace (-3, 3, 100)
ys = np. array ( [ f_true_pol (x) + np. random . randn () *1 for x in xs] )
```

## Visualization of the Dataset

```python
plt.plot(xs,ys,'o', label = 'Artificial Data')
plt.plot(xs,f_true_pol(xs),label = 'Original curve')
plt.xlabel('xs')
plt.ylabel('ys')
plt.legend()
plt.grid()
plt.savefig('Plots/Data_curve.png',format = 'png', bbox_inches = 'tight')
```

```python
### Hypothesis function
def h_pol(x,theta):
    return theta [0] + theta[1]*x + theta[2]*x**2

### Cost function
def J_pol(theta, xs, ys):
    return sum((h_pol(xs, theta) - ys)**2)/(2*len(ys))

### Derivative of the coost function
def gradient_pol (i, theta , xs , ys):
    return sum((h_pol(xs, theta) - ys)*xs**i)/(len(ys))

### Plotting function
def print_modelo_pol (theta , xs , ys, grau = 1):
    plt.plot(xs, h_pol(xs,theta), '--', label = 'Prediction')
    plt.plot(xs,ys,'o', label = 'Artificial data')
    plt.plot(xs,f_true_pol(xs),'-.',label = 'Original curve')
    plt.xlabel('xs')
    plt.ylabel('ys')
    plt.legend()
    plt.grid()
    plt.savefig('Plots/Poli.png',format = 'png', bbox_inches = 'tight')
```

## Gradient Descent for the polynomial curve

```python
def grad_desc_pol(alpha, xs, ys,Silent_Val = True, Silent_J = True):

    theta = np.zeros(3)
    erro  = 1
    T =np.zeros(3)
    count = 0
    J_count = [J_pol(theta, xs, ys)]
    theta_count = [theta]

    while count < 5000:
        for i in range(len(theta)):
            T[i] = gradient_pol(i, theta , xs , ys)
        theta = theta - alpha*T
        erro = np.linalg.norm(T)
        count = count + 1
        J_count.append(J_pol(theta, xs, ys))
        theta_count.append(theta)
        
    if Silent_Val == False:
        print_modelo_pol (theta , xs , ys)

        
    if Silent_J == False:
        plt.figure()
        plt.loglog(np.linspace(1,5000,len(J_count)),J_count)
        plt.xlabel('Epoch')
        plt.ylabel('Cost function')
        plt.grid()
        plt.savefig('Plots/J_poli.png',format = 'png', bbox_inches = 'tight')
    
    return theta,J_count,theta_count
```

```python
theta,J_count,theta_count = grad_desc_pol(0.1, xs, ys,Silent_Val = False, Silent_J = False)
```

## Multivariable regression

#### Function

```python
def f_true_mult(x):
    return 2+0.8*x[0]+3*x[1]
```

#### Artificial data

```python
xs = np. linspace (-3, 3, 100)
zs = np. linspace (-3, 3, 100)
```

```python
ys = f_true_mult ([xs,zs]) +[0.5*np.random.randn() for i in range(len(xs))]
```

```python
### Hyphotesis function
def h_mult(x,theta):
    return theta [0] + theta[1:]@x

### Cost function
def J_mult(theta, x, ys):
    return sum((h_mult(x, theta) - ys)**2)/(2*len(ys))

### Derivative of the cost function
def gradient_mult (i, theta , xs , ys):
    if i == 0:
        return sum((h_mult(xs, theta) - ys))/(len(ys))
    if i == 1:
        return sum((h_mult(xs, theta) - ys)*xs[0])/(len(ys))
    if i == 2:
        return sum((h_mult(xs, theta) - ys)*xs[1])/(len(ys))
```

```python
def grad_desc_mult(alpha, xs, ys, Silent_J = True):

    theta = np.zeros(3)
    erro  = 1
    T =np.zeros(3)
    count = 0
    J_count = [J_mult(theta, xs, ys)]
    theta_count = [theta]

    while count < 5000:
        for i in range(len(theta)):
            T[i] = gradient_mult(i, theta , xs , ys)
        theta = theta - alpha*T
        erro = np.linalg.norm(T)
        count = count + 1
        J_count.append(J_mult(theta, xs, ys))
        theta_count.append(theta)

        
    if Silent_J == False:
        plt.figure()
        plt.loglog(np.linspace(1,5000,len(J_count)),J_count)
        plt.xlabel('Epoch')
        plt.ylabel('Cost function')
        plt.grid()
        plt.savefig('Plots/Multi.png',format = 'png', bbox_inches = 'tight')
        
    
    return theta,J_count,theta_count
```

```python
theta,J_count,theta_count = grad_desc_mult(0.2, np.array([xs,zs]), ys, Silent_J = False)
```
