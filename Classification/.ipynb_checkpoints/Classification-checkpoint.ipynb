## Auxiliar functions

```python
# Sigmoid function 
def sigmoid (z):
    return 1/(1 + np.exp(-z))

# Hypothesis function
def h(x, theta ):
    return sigmoid(theta[0] + theta[1]*x)

# Binary cross-entropy
def cost(h,y):
    return -y*np.log(h)-(1-y)*np.log(1-h)

# Loss function
def J(theta, xs, ys):
    return sum(cost(h(xs,theta),ys))/(2*len(ys))

### Derivative of the loss function
def gradient (i, theta , xs , ys):
    return sum((h(xs, theta) - ys)*xs**i)/(len(ys))
```

#### Plot function

```python
def print_modelo(theta , xs , ys, grau = 1):
    plt.plot(xs, h(xs,theta), 'x', label = 'Prediction')
    plt.plot(xs,ys,'o', label = 'Ground Truth')
    plt.xlabel('xs')
    plt.ylabel('ys')
    plt.legend()
    plt.grid()
    plt.savefig('Plots/Poli.png',format = 'png', bbox_inches = 'tight')
```

## Gradient Descent Algorithm

```python
def grad_desc(alpha, xs, ys,Silent_Val = True, Silent_J = True):

    theta = np.zeros(2)
    erro  = 1
    T =np.zeros(2)
    count = 0
    J_count = [J(theta, xs, ys)]
    theta_count = [theta]
    acc = []
    front = []

    while count < 5000:
        for i in range(len(theta)):
            T[i] = gradient (i, theta , xs , ys)
            
        theta = theta - alpha*T
        erro = np.linalg.norm(T)
        count = count + 1
        J_count.append(J(theta, xs, ys))
        theta_count.append(theta)
        acc.append(np.sum(np.round(h(xs,theta)) == ys)/len(ys))
        front.append((h(xs,theta)[np.round(h(xs,theta)) == 1].min() + h(xs,theta)[np.round(h(xs,theta)) == 0].max())/2)


    if Silent_Val == False:
        print_modelo (theta , xs , ys)

        
    if Silent_J == False:
        plt.figure()
        plt.loglog(np.linspace(1,5000,len(J_count)),J_count)
        plt.xlabel('Época')
        plt.ylabel('Função Custo (J)')
        plt.grid()
        plt.savefig('J_1.png',format = 'png', bbox_inches = 'tight')
    
    return theta,J_count,theta_count, acc, front
```

## Learning rate = 0.1

```python
theta,J_count,Theta_count,acc, front = grad_desc(0.1, xs, ys,Silent_Val = False)
```
